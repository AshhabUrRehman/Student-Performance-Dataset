```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student Performance Dataset.csv")
```

# Drop Student_id Column

```python
df = df.drop("student_id", axis = 1)
```

```python
print(df.head())
```

# Change study_time

```python
df["study_time"] = df["study_time"].replace(15,16)
```

```python
print(df.head())
```

# Gender Distribution

```python
plt.figure(figsize =(5,5))
ax = sns.countplot(data = df, x = "gender")
plt.title("Gender Distribution")
ax.bar_label(ax.containers[0])
plt.show()
```

### From the above chart we have analised that the number of male and female in the data are equal

```python
gb = df.groupby("gender").agg({'math_score':'mean','science_score':'mean','english_score':'mean','overall_score':'mean'})
print(gb)
```

```python
plt.figure(figsize =(5,3))
sns.heatmap(gb,annot = True)
plt.title("Gender Relation On Score")
plt.show()
```

```python
gb1 = df.groupby("study_time").agg({'math_score':'mean','science_score':'mean','english_score':'mean','overall_score':'mean'})
print(gb1)
```

```python
plt.figure(figsize =(5,3))
sns.heatmap(gb1,annot = True)
plt.title(" Study Time Relation On Scores")
plt.show()
```

### From both of the above charts, we concluded that study time has no impact on children's scores, but their scores are affected by gender.

```python
gb2 = df.groupby("extracurricular").agg({'math_score':'mean','science_score':'mean','english_score':'mean','overall_score':'mean'})
print(gb2)
```

```python
plt.figure(figsize =(5,3))
sns.heatmap(gb2,annot = True)
plt.title(" Extra Curricular Relation On Scores")
plt.show()
```
